```python
pickle_in = open("RAW DATA.pickle","rb")
Base_Data = pickle.load(pickle_in)
pickle_in.close()
```

```python
print(Base_Data.columns)
```

```python
Base_Data.tail(50)
```

```python
Base_Data['file'] = Base_Data['file'].str.lower()
Base_Data['body'] = Base_Data['body'].str.lower()
```

```python
Base_Data['file'] = Base_Data['file'].str.replace(r'\W+',"")
```

```python
Base_Data['file'] = Base_Data['file'].str.replace(r'^0-9',"")
Base_Data['body'] = Base_Data['body'].str.replace(r'^0-9',"")
```

```python
Base_Data['file'] = Base_Data['file'].str.replace(r'[^\w\s]|_', '', regex=True)
```

```python
Base_Data['file'] = Base_Data['file'].str.replace('\d+', '')
```

```python
Base_Data['file'] = Base_Data['file'].str.replace('psentmail','', regex= True)
Base_Data['file'] = Base_Data['file'].str.replace('sentitems','', regex= True)
```
